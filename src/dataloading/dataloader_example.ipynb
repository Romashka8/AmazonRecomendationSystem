```python
from dataloader import AmazonReviewsLoader
```

### Создание экземпляра класса с настройками по умолчанию

```python
# настроим пути входящих и выходящих файлов
input_path = "/home/roman/Документы/AmazonRecomendationSystem/data/raw/all_beauty/All_Beauty.jsonl"
output_path = "/home/roman/Документы/AmazonRecomendationSystem/data/interim/"
```

```python
# Создаем экземпляр класса, укажем свой размер выборки и свое семя случайной генерации
loader = AmazonReviewsLoader(input_path, sample_size=1000, seed=19)
```

### Получение выборки данных

```python
# Получаем выборку данных
sample_df = loader.get_sample()

# Просматриваем первые пять строк выборки
sample_df.head()
```

### Получение всех данных

```python
# Получаем все данные
all_data_df = loader.get_all()

# Просматриваем первые пять строк всех данных
all_data_df.head()
```

### Сохранение выборки данных в CSV-файл

```python
# Сохраняем выборку данных в CSV-файл
# Обратите внимание на то, что в файл будет сохранена последняя сгенерированная выборка(в этом случае это весь выгруженный датасет целиком)
loader.save_sample(output_path, "csv")
```

### Изменение семени и создание новой выборки

```python
# Меняем семя для генерации новых случайных индексов
# Сначала изменим размер выборки
loader.sample_size = 1000
loader.reset_seed(42)

# Создаем новую выборку с новым семенем
new_sample_df = loader.get_sample(create_new=True)

# Просматриваем первые пять строк новой выборки
new_sample_df.head()
```

### Получение выборки с сбросом индекса

```python
# Получаем выборку с перезагрузкой индекса
reindexed_sample_df = loader.get_sample(reset_index=True)

# Просматриваем первые пять строк выборки с перезагруженным индексом
reindexed_sample_df.head()
```
